El modelo se expresa como:

El objetivo de la regresi贸n lineal es encontrar los valores 贸ptimos de \( beta_0 \) y \( beta_1 \) que minimicen la diferencia entre los valores predichos y los valores reales.

# Intercepto y Pendiente en Regresi贸n Lineal

En la **regresi贸n lineal**, la relaci贸n entre una variable dependiente \( y \) y una variable independiente \( X \) se modela con la ecuaci贸n:

$$
y = \beta_0 + \beta_1 X + \epsilon
$$

Donde:

- \( y \) es la variable dependiente (lo que queremos predecir).
- \( X \) es la variable independiente (predictor).
- \( \beta_0 \) es el **intercepto**.
- \( \beta_1 \) es la **pendiente**.
- \( \epsilon \) es el **error**.

---

## **1锔 Intercepto \( beta_0 \)**

El **intercepto** representa el valor de \( y \) cuando \( X = 0 \). Es el punto donde la l铆nea de regresi贸n cruza el eje \( y \).

Si el modelo es:

$$
y = 5 + 2X
$$

- Cuando \( X = 0 \), entonces \( y = 5 \).
- Esto significa que si la variable independiente es cero, la variable dependiente tendr谩 un valor de **5**.

### **Ejemplo Real**
Si estamos prediciendo el salario en funci贸n de los a帽os de experiencia:

$$
\text{Salario} = 20,000 + 3,000 \times \text{A帽os de experiencia}
$$

Aqu铆:
- **El intercepto \( 20,000 \)** significa que una persona con **0 a帽os de experiencia** tendr谩 un salario base de **20,000**.

---

## **2锔 Pendiente \( beta_1 \)**

La **pendiente** indica cu谩nto cambia \( y \) cuando \( X \) aumenta en una unidad.

$$
\beta_1 = \frac{\Delta y}{\Delta X}
$$

Si el modelo es:

$$
y = 5 + 2X
$$

- La pendiente es **2**, lo que significa que **por cada aumento de 1 en \( X \)**, \( y \) **aumenta en 2**.

### **Ejemplo Real**
En el modelo de salario:

$$
\text{Salario} = 20,000 + 3,000 \times \text{A帽os de experiencia}
$$

- **La pendiente \( 3,000 \)** significa que **por cada a帽o adicional de experiencia**, el salario esperado **aumenta en 3,000**.

---


## Funci贸n de Costo en Regresi贸n Lineal

Para encontrar los coeficientes 贸ptimos \( beta_0 \) y \( beta_1 \), utilizamos una funci贸n de costo que mida el error entre las predicciones del modelo y los valores reales.

La funci贸n de costo m谩s com煤n es la de **m铆nimos cuadrados**, que se define como:

$$ J(\mathbf{\beta}) = \frac{1}{n} \sum_{i=1}^{n} (y_i - (\beta_0 + \beta_1 X_i))^2 $$

Esta funci贸n representa el error cuadr谩tico medio entre los valores observados \( y_i \) y los valores predichos por el modelo 
$$ y_i - (\beta_0 + \beta_1 X_i) $$

Minimizar esta funci贸n nos permite encontrar los valores 贸ptimos de $$ \mathbf{\beta} $$ que mejor se ajustan a los datos.

## Soluci贸n por M铆nimos Cuadrados

La minimizaci贸n de la funci贸n de costo se puede hacer de manera anal铆tica utilizando la ecuaci贸n normal:

$$ \mathbf{\beta} = (X^T X)^{-1} X^T y $$

El vector de coeficientes beta es igual al inverso del producto de la transpuesta de X por X, multiplicado por la transpuesta de X y luego por el vector de respuestas y

Donde:
- \( X \) es la matriz de dise帽o que contiene las variables independientes (con una columna de unos para el intercepto).
- \( y \) es el vector de valores observados.
- \( beta \) es el vector de coeficientes estimados.

Este m茅todo nos da una soluci贸n cerrada para los coeficientes de la regresi贸n. Sin embargo, tambi茅n podemos resolver el problema utilizando **optimizaci贸n num茅rica** con `scipy.optimize.minimize`.

## Ejercicio 1: Implementaci贸n de Regresi贸n Lineal con scipy.optimize.minimize

- Define la funci贸n de costo

```python
import numpy as np
from scipy.optimize import minimize

# Generar datos sint茅ticos
np.random.seed(42)
X = np.random.uniform(-np.pi * 3 / 4, np.pi, 100)
X.sort()
y = np.sin(X) + np.random.uniform(-0.5, 0.5, 100)

# Agregar columna de unos para el intercepto
X_b = np.c_[np.ones((X.shape[0], 1)), X]

# Definir la funci贸n de costo (Error cuadr谩tico medio)
def criterio_opt(beta, X, y):
    y_pred = X @ beta  # Predicci贸n de y con los par谩metros beta
    error = y - y_pred  # Residual (diferencia entre valores reales y predichos)
    return np.mean(error**2)  # Retornar el error cuadr谩tico medio (MSE)

# Estimaci贸n de par谩metros con optimizaci贸n
beta_inicial = np.zeros(X_b.shape[1])  # Inicializaci贸n de beta en ceros
resultado = minimize(criterio_opt, beta_inicial, args=(X_b, y), method='BFGS')

# Coeficientes 贸ptimos obtenidos
beta_optimo = resultado.x
print(f'Coeficientes 贸ptimos: Intercepto = {beta_optimo[0]:.4f}, Pendiente = {beta_optimo[1]:.7f}')
print(f'Ecuaci贸n y = {beta_optimo[0]:.2f} + {beta_optimo[1]:.7f}X')
```

## Ejercicio 2: Implementaci贸n de Regresi贸n Lineal con scipy.optimize.minimize abstrayendo el concepto en una clase. 

Implementa la siguiente clase `RegresionLineal`, donde debes:

1. **Definir la funci贸n de costo** basada en m铆nimos cuadrados.
2. **Implementar el m茅todo `fit()` usando `scipy.optimize.minimize`**. El m茅todo debe encontrar los coeficientes 贸ptimos.
3. **Incluir un m茅todo `visualizar()` para graficar la recta ajustada**.
4. **Implementar un m茅todo `predecir()`** para realizar predicciones.
5. **Definir un m茅todo `mostrar_ecuacion()`** para imprimir la ecuaci贸n ajustada.
6. **Comparar los resultados con `sklearn.linear_model.LinearRegression`**.
7. Prueba todos los m茅todos con X e y definidos anterioremente. 

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize
from sklearn.linear_model import LinearRegression

class RegresionLineal:
    def __init__(self, X, y):
        self.X = np.c_[np.ones((X.shape[0], 1)), X]  # Agregar columna de 1s para el intercepto
        self.y = y
        self.theta = None  # Inicialmente no tenemos coeficientes

    def funcion_costo(self, theta):
        """Funci贸n de costo basada en m铆nimos cuadrados (error cuadr谩tico medio)."""
        y_pred = self.X @ theta  # Predicci贸n
        error = self.y - y_pred  # Diferencia entre valores reales y predichos
        return np.mean(error ** 2)  # Retornar el error cuadr谩tico medio (MSE)

    def fit(self):
        """Ajusta los coeficientes usando scipy.optimize.minimize."""
        theta_inicial = np.zeros(self.X.shape[1])  # Inicializar los coeficientes en cero
        resultado = minimize(self.funcion_costo, theta_inicial, method='BFGS')
        self.theta = resultado.x  # Guardar los coeficientes 贸ptimos

    def visualizar(self):
        """Grafica los datos junto con la recta ajustada."""
        if self.theta is None:
            print("Primero debes ajustar el modelo con fit().")
            return

        plt.scatter(self.X[:, 1], self.y, label="Datos", color='blue', alpha=0.6)
        X_graf = np.linspace(self.X[:, 1].min(), self.X[:, 1].max(), 100)
        y_graf = self.theta[0] + self.theta[1] * X_graf
        plt.plot(X_graf, y_graf, color='red', label="Recta Ajustada")
        plt.xlabel("X")
        plt.ylabel("y")
        plt.legend()
        plt.title("Regresi贸n Lineal - Ajuste de la Recta")
        plt.show()

    def predecir(self, X_nuevo):
        """Realiza una predicci贸n para nuevos valores de entrada."""
        X_nuevo = np.c_[np.ones((X_nuevo.shape[0], 1)), X_nuevo]  # Agregar columna de 1s
        return X_nuevo @ self.theta  # Predicci贸n usando los coeficientes ajustados

    def mostrar_ecuacion(self):
        """Imprime la ecuaci贸n de la regresi贸n lineal."""
        if self.theta is None:
            print("Primero debes ajustar el modelo con fit().")
        else:
            print(f"Ecuaci贸n de la regresi贸n: y = {self.theta[0]:.4f} + {self.theta[1]:.7f}X")

# Comparaci贸n con sklearn
modelo_sklearn = LinearRegression()
modelo_sklearn.fit(X.reshape(-1, 1), y)
print(f"Comparaci贸n con sklearn:")
print(f"Intercepto: {modelo_sklearn.intercept_:.4f}, Pendiente: {modelo_sklearn.coef_[0]:.7f}")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize
from sklearn.linear_model import LinearRegression

#SALARY PREDICTION DATASET
# Cargar el archivo CSV
def cargar_datos(ruta):
    df = pd.read_csv(ruta)
    return df.dropna()  # Eliminar filas con valores faltantes

# Definir la clase de Regresi贸n Lineal
class RegresionLineal:
    def __init__(self, X, y):
        self.X = np.c_[np.ones((X.shape[0], 1)), X]
        self.y = y
        self.theta = None

    def funcion_costo(self, theta):
        y_pred = self.X @ theta
        error = self.y - y_pred
        return np.mean(error ** 2)

    def fit(self):
        theta_inicial = np.zeros(self.X.shape[1])
        resultado = minimize(self.funcion_costo, theta_inicial, method='BFGS')
        self.theta = resultado.x

    def visualizar(self):
        if self.theta is None:
            print("Primero debes ajustar el modelo con fit().")
            return
        
        plt.scatter(self.X[:, 1], self.y, label="Datos", color='blue', alpha=0.6)
        X_graf = np.linspace(self.X[:, 1].min(), self.X[:, 1].max(), 100)
        y_graf = self.theta[0] + self.theta[1] * X_graf
        plt.plot(X_graf, y_graf, color='red', label="Recta Ajustada")
        plt.xlabel("Years of Experience")
        plt.ylabel("Salary")
        plt.legend()
        plt.title("Regresi贸n Lineal - Ajuste de la Recta")
        plt.show()

    def predecir(self, X_nuevo):
        X_nuevo = np.c_[np.ones((X_nuevo.shape[0], 1)), X_nuevo]
        return X_nuevo @ self.theta

    def mostrar_ecuacion(self):
        if self.theta is None:
            print("Primero debes ajustar el modelo con fit().")
        else:
            print(f"Ecuaci贸n de la regresi贸n: Salary = {self.theta[0]:.4f} + {self.theta[1]:.7f} * (Years of Experience)")

# Ruta del archivo CSV
ruta_Salary = "Salary Data.csv"

df = cargar_datos(ruta_Salary)
X = df["Years of Experience"].values.reshape(-1, 1)
y = df["Salary"].values

# Entrenar el modelo personalizado
modelo = RegresionLineal(X, y)
modelo.fit()
modelo.mostrar_ecuacion()
modelo.visualizar()

# Comparaci贸n con sklearn
modelo_sklearn = LinearRegression()
modelo_sklearn.fit(X, y)
print(f"Comparaci贸n con sklearn:")
print(f"Intercepto: {modelo_sklearn.intercept_:.4f}, Pendiente: {modelo_sklearn.coef_[0]:.7f}")

# Realizar predicciones con nuevos datos
X_nuevo = np.array([[5], [10], [15]])
y_pred = modelo.predecir(X_nuevo)
print("Predicciones para nuevos valores de Years of Experience:")
for i in range(len(X_nuevo)):
    print(f"Para {X_nuevo[i][0]} a帽os de experiencia, salario estimado: {y_pred[i]:.2f}")


#AUTO MPG DATASET
# Cargar los datos desde el archivo CSV
data = pd.read_csv("Auto MPG Data.csv")

# Eliminar filas con valores faltantes en la columna 'horsepower'
data = data[data['horsepower'] != '?']

# Convertir 'horsepower' a tipo num茅rico
data['horsepower'] = data['horsepower'].astype(float)

# Preparar los datos de entrada X y salida y
X = data[['cylinders', 'displacement', 'horsepower', 'weight', 'acceleration']]
y = data['mpg']

# Instanciar y entrenar un modelo de regresi贸n lineal personalizado
class RegresionLineal:
    def __init__(self):
        self.coef_ = None
        self.intercept_ = None

    def fit(self, X, y):
        ones = np.ones((X.shape[0], 1))
        X = np.concatenate((ones, X), axis=1)
        self.coef_ = np.linalg.inv(X.T @ X) @ X.T @ y
        self.intercept_ = self.coef_[0]
        self.coef_ = self.coef_[1:]

    def predict(self, X):
        return self.intercept_ + np.dot(X, self.coef_)

model_custom = RegresionLineal()
model_custom.fit(X.to_numpy(), y.to_numpy())

# Instanciar y entrenar un modelo de regresi贸n lineal de sklearn
model_sklearn = LinearRegression()
model_sklearn.fit(X, y)

# Comparar los resultados con el modelo de sklearn
print("Coeficientes (custom):", model_custom.coef_)
print("Intercept (custom):", model_custom.intercept_)
print("Coeficientes (sklearn):", model_sklearn.coef_)
print("Intercept (sklearn):", model_sklearn.intercept_)

# Visualizar la relaci贸n entre las variables
plt.scatter(data['horsepower'], y, color='blue')
plt.plot(data['horsepower'], model_custom.predict(X.to_numpy()), color='red', linewidth=2)
plt.xlabel('Horsepower')
plt.ylabel('MPG')
plt.title('Relaci贸n entre Horsepower y MPG')
plt.show()

# Mostrar la ecuaci贸n de regresi贸n obtenida
print("Ecuaci贸n de regresi贸n (custom):")
print("MPG = {:.2f} + {:.2f}*cylinders + {:.2f}*displacement + {:.2f}*horsepower + {:.2f}*weight + {:.2f}*acceleration".format(model_custom.intercept_, *model_custom.coef_))

# Realizar predicciones con datos nuevos
new_data = [[4, 140, 80, 2000, 15]]  # Ejemplo de datos nuevos
prediction_custom = model_custom.predict(new_data)
prediction_sklearn = model_sklearn.predict(new_data)

print("Predicci贸n (custom):", prediction_custom)
print("Predicci贸n (sklearn):", prediction_sklearn)
```
